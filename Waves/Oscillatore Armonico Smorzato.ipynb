```python
# Risoluzione numerica ODE Oscillatore Armonico Smorzato
import numpy as np
from matplotlib import pyplot as plt

m = 1
k = 1
x0 = 1
v0 = 0
a0 = 0
t = 0
tmax = 60
lm = 0.5 # coefficiente di smorzamento (lm = 2 si ottiene smor critico)
dt = 0.1

ris_x = []
ris_v = []
ris_a = []
ris_t = [0]

# Metodo di Runge-Kutta del IV ordine (per risoluzione di ODE)
def calcolo (x,v,a):
    # parametro 1
    k1 = v*dt
    j1 = a*dt
    vm1 = v+j1/2
    am1 = -k*(x+k1/2)/m - lm*(v+j1/2)/m 
    # parametro 2
    k2 = vm1*dt
    j2 = am1*dt
    vm2 = v + j2/2
    am2 = -k*(x+k2/2)/m - lm*(v+j2/2)/m
    # parametro 3
    k3 = vm2*dt
    j3 = am2*dt
    vf = v+j3
    af = -k*(x+k3)/m - lm*(v+j3)/m
    # parametro 4
    k4 = vf*dt
    j4 = af*dt
    # risultati
    x = x + (k1+2*k2+2*k3+k4)/6
    v = v+(j1+2*j2+2*j3+j4)/6
    a = -k*x/m - lm*v/m
    return x,v,a

x,v,a = calcolo(x0,v0,a0)
t += dt
ris_x.append(x)
ris_v.append(v)
ris_a.append(a)

while (t <= tmax):
    ris_t.append(t)
    x,v,a = calcolo(x,v,a)
    ris_x.append(x)
    ris_v.append(v)
    ris_a.append(a)
    t += dt

print(ris_x)
print(ris_v)
print(ris_a)

plt.xlabel('Tempo', loc='right')
plt.ylabel('Posizione', loc="top")
plt.grid(linestyle='--', linewidth=0.4)
plt.plot(ris_t,ris_x, "black")
plt.show()

plt.xlabel('tempo', loc='right')
plt.ylabel('Velocit', loc="top")
plt.grid(linestyle='--', linewidth=0.4)
plt.plot(ris_t,ris_v, "red")
plt.show()

plt.xlabel('Tempo', loc='right')
plt.ylabel('Accelerazione', loc="top")
plt.grid(linestyle='--', linewidth=0.4)
plt.plot(ris_t,ris_a, "blue")
plt.show()
    
    
    
```
